```python
import matplotlib.pyplot as plt
from skimage import io, segmentation
from skimage.util import img_as_float

# Load image
image = img_as_float(io.imread('input.jpg'))

# Apply Felzenszwalb segmentation
segments = segmentation.felzenszwalb(
    image, 
    scale=500,       # k parameter (higher=k means larger segments)
    sigma=0.8,       # Gaussian smoothing pre-processing
    min_size=50      # Minimum component size (enforced post-processing)
)

# Show results
fig, ax = plt.subplots(1, 2, figsize=(12,6))
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[1].imshow(segmentation.mark_boundaries(image, segments))
ax[1].set_title('Segmentation Boundaries')
plt.show()
```
